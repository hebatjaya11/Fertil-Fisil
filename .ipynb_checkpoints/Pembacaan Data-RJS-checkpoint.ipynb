```python
#Membuat fungsi untuk mengambil nili k-inf
def readfile(filedir):
    keyword = "Transport: k-inf ="
    kinf = [] 
    f = open(filedir, "r")
    for line in f:
        if keyword in line:
            kval = line.split()
            kinf.append(float(kval[3]))
    return kinf

# FUngsi untuk mengekstrak nilai suhu pada nama file
def extract_number_from_filename(filename):
    # Menggunakan regex untuk mengekstrak angka setelah awalan yang konsisten
    match = re.search(r'_(\d+(\.\d+)?)_', filename)  # Mencocokkan angka setelah _
    if match:
        number = float(match.group(1))  # Mengembalikan angka sebagai float, grup (1) berisi angka
        return number
    return 0

mycwd = os.getcwd()
```

```python
tipe = ["Fertil-Fisil", "Fisil-Fertil", "Homogen"]
Npin = np.arange(10, 18, 1)
koef = ["FTC", "MTC", "VC"]
varpm = np.arange(5,55,5) 
var_suhu_Fuel = np.linspace(800, 1000, 5) # TF nominal = 900 K
var_suhu_Moderator = np.linspace(541, 601, 7) # Tmode nominal = 571 K
Waktu_Burnup_thorium = [0, 1, 8, 30, 240, 1200, 1980, 2520, 3240, 3960]
Setpoint = np.linspace(1, len(Waktu_Burnup_thorium), 10)

x = np.linspace(0.01, 0, 10)  # fraksi void dari 1% sampai 0%
Vfrac = x*100
rho_l = 0.72998
rho_g = 0.044553
rhoM_v = np.round((1-x)*rho_l + x*rho_g, 6)


mycwd = os.getcwd()

#list_kinf_ferfil = []
list_kinf_thorium = []
for tp in tipe:
    for Np in Npin:
        for alpha in koef:
            for var in varpm:
                if alpha == "FTC":
                    path_th = f"{mycwd}/{tp}/{Np}/{alpha}/{var}"
                    os.chdir(path_th)
                    file_list = Path(".").glob("*.out")
                    list_file = [str(file) for file in file_list]
                    sorted_files = sorted(list_file, key=extract_number_from_filename) #udah urut
                    #print(sorted_files)
                    ferfil_FTC_row = len(sorted_files)
                    for i in range (ferfil_FTC_row):
                        list_kinf_thorium.append(readfile(sorted_files[i]))
                        if i == ferfil_FTC_row-1:
                            for i in range (10-ferfil_FTC_row):
                                list_padding = [np.nan]*10
                                list_kinf_thorium.append(list_padding)
                                
                    os.chdir(mycwd)
                if alpha == "MTC":
                    path_th = f"{mycwd}/{tp}/{Np}/{alpha}/{var}"
                    os.chdir(path_th)
                    file_list = Path(".").glob("*.out")
                    list_file = [str(file) for file in file_list]
                    sorted_files = sorted(list_file, key=extract_number_from_filename) #udah urut
                    ferfil_MTC_row = len(sorted_files)
                    for i in range (ferfil_MTC_row):
                        list_kinf_thorium.append(readfile(sorted_files[i]))
                        if i == ferfil_MTC_row-1:
                            for i in range (10-ferfil_MTC_row):
                                list_padding = [np.nan]*10
                                list_kinf_thorium.append(list_padding)
                    os.chdir(mycwd)
                if alpha == "VC": #cek urutan plot data apakah sudah sesuai, soalya ini kan rhomnya berkebalikan dengan besar persentase void
                    path_th = f"{mycwd}/{tp}/{Np}/{alpha}/{var}"
                    os.chdir(path_th)
                    file_list = Path(".").glob("*.out")
                    list_file = [str(file) for file in file_list]
                    sorted_files = sorted(list_file, key=extract_number_from_filename) #udah urut
                    # print(sorted_files)
                    # if Np == 16:
                    #     if var == 25:
                    #         print(sorted_files)
                    ferfil_VC_row = len(sorted_files)
                    for i in range (ferfil_VC_row):
                        list_kinf_thorium.append(readfile(sorted_files[i]))
                    os.chdir(mycwd)
os.chdir(mycwd)
array_kinf_thorium = np.array(list_kinf_thorium).reshape(len(tipe), len(Npin), len(koef),len(varpm), len(Vfrac), len(Waktu_Burnup_thorium))

def Nilai_kinf (tp, Np, alpha, var, suhu_atau_rhoM_V, Sp):
    indeks_tipe = tipe.index(tp)
    indeks_Npin = np.where(Npin==Np)[0][0]
    indeks_alpha = koef.index(alpha)
    if alpha == "FTC":
        indeks_suhu_atau_rhoM_V = np.where(var_suhu_Fuel==suhu_atau_rhoM_V)[0][0]
    elif alpha == "MTC":
        indeks_suhu_atau_rhoM_V = np.where(var_suhu_Moderator==suhu_atau_rhoM_V)[0][0]
    elif alpha == "VC":
        #aku ingin ketika rhomV diinput akan dikeluarkan indeks file outputnya
        indeks_suhu_atau_rhoM_V = np.where(rhoM_v==suhu_atau_rhoM_V)[0][0]
    indeks_varpm = np.where(varpm==var)[0][0]
    indeks_Setpoint = np.where(Setpoint==Sp)[0][0]
    kinf = array_kinf_thorium[indeks_tipe][indeks_Npin][indeks_alpha][indeks_varpm][indeks_suhu_atau_rhoM_V][indeks_Setpoint]
    #Hasil = f"Nilai kinf pada Tipe = {tp}, pin = {Np}, alpha = {alpha}, Suhu = {suhu_atau_rhoM_V}, dan Setpoint ke-{Sp} adalah = {kinf}"
    return print(kinf)

```

```python
# Membuat fungsi regresi
def linear_regression(x, y):
    # Ubah data menjadi array numpy dan reshape jika perlu
    x = np.array(x).reshape(-1, 1)
    y = np.array(y)

    # Buat model regresi linear
    model = LinearRegression()

    # Latih model dengan data
    model.fit(x, y)

    # Ambil koefisien dan intercept
    slope = model.coef_[0]
    intercept = model.intercept_

    # Buat persamaan garis dalam bentuk string
    #equation = f'y = {slope:.2f}x + {intercept:.2f}'

    return slope, intercept
```
